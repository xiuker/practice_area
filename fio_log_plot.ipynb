```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', **{'family':'Microsoft YaHei, SimHei'})
import os
```

```python
filelist = [i for i in os.listdir(os.getcwd()) if '.log' in i]
```

```python
X = np.loadtxt(filelist[0],delimiter=',')
```

```python
X[:,-1] = np.cumsum(X[:,-1] / 1000000)
X[:,1] = X[:,1] / 1000
X[:,0] = X[:,0] / 1024
```

```python
plt.figure(figsize=(10,5))
plt.plot(X[:,0],X[:,1],marker='.',label='mbps')
plt.legend()
#plt.axis([-5,260,100,1500])
plt.xlabel('/G')
plt.ylabel('/M')
plt.title('-')
plt.grid(True)
plt.savefig('-.png',dpi=1080)
```

```python
plt.figure(figsize=(10,5))
plt.plot(X[:,-1],X[:,1],marker='.',label='mbps')
plt.legend()
#plt.axis([-5,260,100,1500])
plt.xlabel('/s')
plt.ylabel('/M')
plt.title('-')
plt.grid(True)
#plt.show()
plt.savefig('-.png',dpi=1080)
```
